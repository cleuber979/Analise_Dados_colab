```python
import pandas as pd
```

```python
filename = '/content/drive/MyDrive/iris.csv'
df=pd.read_csv(filename, index_col=0)
df.head()
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('iris.db')
```

```python
df.to_sql(name='flores',con=conn)
```

```python
df_read = pd.read_sql('SELECT * FROM flores', con=conn)
```

```python
df_read
```

```python
query = """
         SELECT  "sepal_length","sepal_width","petal_length","species"
         FROM FLORES;
"""

df_read = pd.read_sql(query,con=conn)
df_read
```

```python
query = """
         SELECT *
         FROM FLORES
         WHERE "species" = "virginica";
"""

df_read = pd.read_sql(query,con=conn)
df_read
```

```python
# fechando o banco de dados
conn.close()
```
